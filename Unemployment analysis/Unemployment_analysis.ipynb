# **Analysing and Visualization**

```python
#Importing the necessary Libraries
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
#Visualizing using Bar graph ( Estimated Unemployment Rate (%))
fig = px.bar(df,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',title='Unemployment Rate in every state using Bar Graph')
fig.update_layout(xaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
#Visualizing using Scatter Plot( Estimated Unemployment Rate (%))
fig = px.scatter(df,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',
            title='Unemployment Rate in every state using ScatterPlot',)
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
#Visualizing using Bar graph ( Estimated Unemployment Rate (%))
fig = px.histogram(df,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',template='plotly',
                  title='Unemployment Rate in every state using Histogram')
fig.update_layout(xaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
#Visualizing using Bar graph (Estimated Employed)
fig = px.bar(df,x='Region',y=" Estimated Employed",color='Region',
            title=' Estimated Employment Rate using Bar graph')
fig.update_layout(xaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
##Visualizing using Boxplot (Estimated Labour Participation Rate (%))
fig = px.box(df,x='Region',y=" Estimated Labour Participation Rate (%)",color='Region',
            title='Estimated Labour Participation Rate (%) in every state using Box Plot',)
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```
